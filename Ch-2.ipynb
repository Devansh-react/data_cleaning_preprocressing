# Select Slow and Fast Pokemons Based on Speed Column Slow Below 25% and Fast Above 75% and Above Mean Speed

```python
df["Speed"].describe()

speed_25 = df["Speed"].quantile(0.25)
speed_75 = df["Speed"].quantile(0.75)
[speed_25, speed_75]
```

```python
df.loc[(df["Speed"] >= speed_75) | (df["Speed"] <= speed_25) ,["Name","Generation","Speed"]].head(10)
```

# Pokemon with Strong Defense and Strong Attack

```python
strong_attack = df["Attack"].quantile(0.80)
strong_defense = df['Defense'].quantile(0.80)

pokemon = df.loc[(df["Attack"]>=strong_attack) & (df["Defense"]>=strong_defense),["Name","Attack","Defense"]].head(5)
pokemon
```

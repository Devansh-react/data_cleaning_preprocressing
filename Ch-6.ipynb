# Data Cleaning and Processing using Pandas

Rerplace the invalid values from the dataset with proper values `season column ` and replace it with `Unknown season`

```python
df["season"].value_counts()
```

```python
temp_df = df.loc[df["season"]== '?',"season"].str.replace('?',"Unkown season")
```

```python
df.loc[df["season"]== '?',"season"] = temp_df
```
